```python
#part 1
#Bank class 
class Bank:
    def __init__(self, ifsc_code = None, bank_name = None, branch_name = None, loc=None):
        self.ifsc_code = ifsc_code
        self.bank_name = bank_name
        self.branch_name = branch_name
        self.loc = loc

    def print_info(self):
        print("IFSC_Code :", self.ifsc_code)
        print("Bank Name :", self.bank_name)
        print("Branch Name :", self.branch_name)
        print("Location :", self.loc)

#customer class
class Customer:
    def __init__(self, customer_id = None, cust_name = None, address = None, contact_details = None):
        self.customer_id = customer_id
        self.cust_name = cust_name
        self.address = address
        self.contact_details = contact_details
    
    def print_info(self):
        print("Customer ID :", self.customer_id)
        print("Customer Name :", self.cust_name)
        print("Address :", self.address)
        print("Contact Details :", self.contact_details)

#Account Class
class Account(Bank):
    def __init__(self, ifsc_code, bank_name, branch_name, location, account_id, cust_obj, balance=0):
        super().__init__(ifsc_code, bank_name, branch_name, location)
        self.account_id = account_id
        self.cust_obj = cust_obj
        self.balance = balance

    def display(self):
        self.print_info()
        print("Account ID :", self.account_id)
        print("Cust Object of Customer :",self.cust_obj)
        print("balance :", self.balance)

    def get_account_info(self):
        print("Customer ID " + str(self.cust_obj.customer_id))
        print("Customer Name " + str(self.cust_obj.cust_name))
        print("Contact Details " + str(self.cust_obj.contact_details))

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposit successful. New balance: {self.balance}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrawal successful. New balance: {self.balance}")
        else:
            print("Insufficient funds. Withdrawal unsuccessful.")

    def get_balance(self):
        return self.balance

#saving account class
class SavingsAccount(Account):
    def __init__(self, ifsc_code, bank_name, branch_name, location, account_id, cust_obj, s_min_balance, balance=0):
        super().__init__(ifsc_code, bank_name, branch_name, location, account_id, cust_obj, balance)
        self.s_min_balance = s_min_balance

    def get_savings_account_info(self):
        self.get_account_info()
        print(f"Minimum Balance: {self.s_min_balance}")
    
    def withdraw(self, amount):
        if self.balance - amount >= self.s_min_balance:
            super().withdraw(amount)
        else:
            print("Withdrawal unsuccessful. Below minimum balance.")

#banking system class from there the application starts working
class BankingSystem:
    def __init__(self):
        # Accepting input from the end user to create respective class objects
        customer_id         = "C001"
        customer_name       = "aditya verma"
        customer_address    = "ABC location xyz"
        contact_details     = "9876543210"

        # # Input passed : Customer ID , Customer name , Customer address , contact details
        customer = Customer(customer_id,customer_name , customer_address, contact_details)

        #ifsc_code, bank_name, branch_name, location
        ifsc_name   = "DEN123"
        bank_name   = "Dena bank"
        branch_name = "Main Branch"
        location    = "Shashtripuram , Agra"

        account_id  = "A002"

        # customer id and account id will be different , 
        # since one customer can have more than one account 
        # Also customer id is used for online banking while account id is used for online payment transactions
        # in doc it is SavigsAccount i believe its mispelled there
        self.savings_account = SavingsAccount(ifsc_name, bank_name, branch_name, location, account_id, customer, 500, 5000)

    def choice_1(self):
        print("\nBank Details:")
        print(f"IFSC Code: {self.savings_account.ifsc_code}")
        print(f"Bank Name: {self.savings_account.bank_name}")
        print(f"Branch Name: {self.savings_account.branch_name}")
        print(f"Location: {self.savings_account.loc}")

    def choice_2(self):
        self.savings_account.get_account_info()

    def choice_3(self):
        balance = self.savings_account.get_balance()
        print(f"Current Balance: {balance}")

    def choice_4(self):
        amount = float(input("Enter the deposit amount: "))
        self.savings_account.deposit(amount)

    def choice_5(self):
        amount = float(input("Enter the withdrawal amount: "))
        self.savings_account.withdraw(amount)
    
bs = BankingSystem()
```
